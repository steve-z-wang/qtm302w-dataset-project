```python
# seperate month and year

month_year = df2['Month'].str.split('-', expand=True).rename({0: 'Month', 1: 'Year'}, axis=1)[['Year', 'Month']]

month_year['Year'] = month_year['Year'].apply(lambda s: '20' + s)

df3 = pd.concat([month_year, df2.drop('Month', axis = 1)], axis = 1)
```

```python
# pivot the dataframe

df4 = pd.melt(df3, id_vars = ['Year', 'Month']).rename({'variable': 'State', 'value': 'Generation'}, axis=1) 
df4['Month'] = pd.Categorical(df4['Month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
```

```python
df4.head()
```

```python
# sort the dataframe by year, state, month

df5 = df4.sort_values(by=['Year', 'State', 'Month'])
```

```python
df5.head()
```

```python
# select data from 2013 to 2020

years = [str(i) for i in range(2013, 2021)]

df6 = df5[df5['Year'].isin(years)]
```

```python
len(df6['State'].unique()) * 8 * 12
```

```python
# calculate month / year percentage for each state

a = pd.pivot_table(df6, index=['Year', 'State'], columns=['Month'], values='Generation', dropna=False)
```

```python
a
```

```python
b = df6.groupby(['Year', 'State']).aggregate({'Generation': 'sum'})
```

```python
b 
```

```python
generation_monthly_percentage = np.divide(a, b)
```

```python
generation_monthly_percentage
```

```python
# import emission data from 2013 to 2020

os.chdir('/Users/stevewang/Documents/Emory/2022-fall/qtm302w/data-set-project/dataset/emission/')

# merging

years = list(range(2013, 2021))

frames = []

for y in years: 
    filename = 'emissions' + str(y) + '.xlsx'
    
    data_df = pd.read_excel(filename, header=1)
    
    data_df.insert(0, "Year", y)

    frames.append(data_df)
    
emission_df = pd.concat(frames).drop([5194, 5195])
```

```python
for c in emission_df.columns:
    print(c)
```

```python
# select and rename columns
emission_df2 = emission_df[['Year', 'Plant Code', 'State', 'Aggregated Fuel Group', 'Generation (kWh)', 'Tons of CO2 Emissions']]
emission_df2 = emission_df2.rename({'Generation (kWh)': 'Generation', 'Tons of CO2 Emissions': 'Emission'}, axis=1)

# only select COAL
emission_df3 = emission_df2[emission_df2['Aggregated Fuel Group'] == 'COAL']

# aggregate the generation and emissions by year and plant code
emission_df4 = emission_df3.groupby(['Year', 'State', 'Plant Code']).aggregate({'Generation': 'sum', 'Emission': 'sum'})
```

```python
emission_df4
```

```python
# combine with the previous data table to calculate the monthly energy generation
```

```python
# calculate the monthly generation for each month

data = []
for p in emission_df4.index: 
    year, state, plant = p[0], p[1], p[2]

    generation = emission_df4.loc[p]['Generation']

    percentage = generation_monthly_percentage.loc[(str(year), state)]

    data.append([year, state, plant] + list(np.dot(generation, percentage)))
```

```python
monthly = pd.DataFrame(data, columns=['Year', 'State', 'Plant Code', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug',
                  'Sep', 'Oct', 'Nov', 'Dec'])

monthly = pd.melt(monthly, id_vars = ['Year', 'State', 'Plant Code']).rename({'variable': 'Month', 'value': 'Generation'}, axis=1)
```

```python
monthly
```

```python
# get the generation to emission function

model = smf.ols(formula='Emission ~ Generation', data=emission_df4)
model = model.fit()
```

```python
model.summary()
```

```python
# use the generation to emission function to get the monthly output for each power plant

monthly['Emission (Predicted)'] = model.predict(monthly)
```
