```python
import os 
import pandas as pd
```

```python
path = os.getcwd()
parent = os.path.dirname(path)
```

```python
os.chdir(parent + '/dataset/')

df = pd.read_excel('EIA923_Schedules_2_3_4_5_M_12_2020_Final_Revision.xlsx', header = 5)
```

```python
cols = ['Plant Id', 'Plant Name', 'Plant State', 'Sector Name', 'Reported\nPrime Mover', 'Reported\nFuel Type Code', 'Respondent\nFrequency'] 
for c in df.columns: 
    if 'Netgen' in c: 
        cols.append(c)

cols
```

```python
df2 = df[cols]
```

```python
mapper = {}
for c in cols: 

    c2 = c

    if '\n' in c: 
        c2 = c2.replace('\n', ' ')

    if 'Netgen' in c:
        c2 = c2.replace('Netgen ', '')[:3]

    mapper[c] = c2
```

```python
mapper
```

```python
df3 = df2.rename(columns = mapper)
```

```python
df4 = df3[df3['Respondent Frequency'] == 'M']
```

```python
df5 = df4[['Plant Id', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']]
```

```python
df5
```

```python
df6 = df5.applymap(lambda x : 0 if x == '.' else x)
```

```python
df6
```

```python
df7 = pd.melt(df6, id_vars = ['Plant Id']).rename({'variable': 'Month', 'value': 'Generation'}, axis=1)
df7
```

```python
df8 = df7.groupby(['Plant Id', 'Month'], as_index=False).aggregate({'Generation': 'sum'}) 
```

```python
df8
```

```python
df7['Month'] = pd.Categorical(df7['Month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
```

```python
df8 = df7.sort_values(by=['Plant Id', 'Month'])
```

```python
df8.head(10)
```

```python
df8['Month'].unique()
```

```python
df9 = pd.pivot_table(df8, index=['Plant Id'], columns=['Month'], values='Generation', dropna=False)
df9
```
